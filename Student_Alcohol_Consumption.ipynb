```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'

df = pd.read_csv(url)

df.head()
```

```python
df_sliced = df.loc[:,'school':'guardian']

df_sliced.head()
```

```python
capital = lambda a: a.capitalize()
```

```python
df_sliced['Mjob'].apply(capital)
df_sliced['Fjob'].apply(capital)
```

```python
df_sliced.tail()
```

```python
df_sliced['Mjob']=df_sliced['Mjob'].apply(capital)
df_sliced['Fjob']=df_sliced['Fjob'].apply(capital)

df_sliced.tail()
```

```python
def majority(age):
  if age>17:
    return True
  else:
    return  False
```

```python
df_sliced['legal_drinker']=df_sliced['age'].apply(majority)

df_sliced.head()
```

```python
def pro10(a):
  if type(a) is int:
    return a*10
  else:
    return a
```

```python
df_sliced.applymap(pro10).head()
```
